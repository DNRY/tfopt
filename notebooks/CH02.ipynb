### 2.2.5 제너레이터(generator)

```````python
def method():
    print("Start method()")
    val = []
    for x in range(0,5):
        val.append( x**2 )
    return val
```````

```````python
out = method()
for i in range(0,5):
    print(out[i])
```````

```````python
def generator():
    print("Start generator()")
    for x in range(0,5):
        yield x**2
```````

```````python
gen = generator()
```````

```````python
for i in range(0,5):
    print(next(gen))
```````

## 2.3 자주 사용되는 파이썬 문법 패턴

### 2.3.1 데이터 타입마다 다른 for loop 스타일

```````python
items = [[1,2], [3,4], [5,6]]
for item in items:
    print(item[0], item[1])
    
for item1, item2 in items:
    print(item1, item2)
```````

```````python
items = [(1,2), (3,4), (5,6)]
for item in items:
    print(item[0], item[1])
    
for item1, item2 in items:
    print(item1, item2)
```````

```````python
info = {'A' : 1, 'B' : 2, 'C' : 3}
for key in info:
    print(key, info[key])
```````

```````python
for key, value in info.items():
    print(key, value)
```````

### 2.3.2 zip가 들어간 for loop

```````python
items1 = [[1,2], [3,4], [5,6]]
items2 = [['A','B'], ['C','D'], ['E','F']]
print(items1)
print(items2)
```````

```````python
for digits, characters in zip(items1, items2):
    print(digits, characters)
```````

### 2.5.1 분포도(Scatter Plot) 그리기

```````python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)

plt.scatter(x, y)
plt.show()
```````

```````python
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area)
plt.show()
```````

```````python
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, alpha=0.5)
plt.show()
```````

```````python
np.random.seed(19680801)

N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = (30 * np.random.rand(N))**2

plt.scatter(x, y, s=area, c=colors, alpha=0.5)
plt.show()
```````

### 2.5.4 여러 그래프를 한 눈에 보기

```````python
x = np.arange(0., 5., 0.2)
y1 = x
y2 = 0.25 * x**2
y3 = np.cos(x)
plt.plot(x,y1)
plt.plot(x,y2)
plt.plot(x,y3)
plt.show()
```````

```````python
plt.plot(x,y1, '-')
plt.plot(x,y2, '--')
plt.plot(x,y3, ':')
plt.show()
```````

```````python
plt.plot(x,y1, 'o-')
plt.plot(x,y2, '--s')
plt.plot(x,y3, ':x')
plt.show()
```````

```````python
plt.plot(x,y1, 'o-')
plt.plot(x,y2, '--s')
plt.plot(x,y3, ':x')
plt.legend(['y=x', 'y=0.25 * x**2', 'y = cos(x)'])
plt.show()
```````

```````python
plt.plot(x, y1, 'o-', x, y2, '--s', x, y3, ':x')
plt.legend(['y=x', 'y=0.25 * x**2', 'y = cos(x)'])
plt.show()
```````

### 2.5.5 그래프 스타일링

```````python
plt.plot(x, y1, 'o-', x, y2, '--s', x, y3, ':x')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend(['$y=x$', '$y=0.25 * x^2$', '$y = cos(x)$'])
plt.show()
```````

```````python
plt.plot(x, y1, 'o-', label=r'$y=x$')
plt.plot(x, y2, '--s', label=r'$y=\frac{1}{4}x^2$')
plt.plot(x, y3, ':x', label=r'$y=cos(x)$')
plt.title('Plot Tutorial')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```````

```````python
plt.subplot(3,1,1)
plt.plot(x, y1, '-ok')
plt.title('Subplots Example')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,2)
plt.plot(x, y2, '--sk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')

plt.subplot(3,1,3)
plt.plot(x, y3, ':xk')
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```````

### 2.5.6 다변수 함수 그래프 그리기

```````python
X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)
```````

```````python
import matplotlib.pyplot as plt
fig = plt.figure(figsize=(5,5))
plt.contour(X, Y, Z, cmap='coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```````

```````python
fig = plt.figure(figsize=(5,5))
plt.contourf(X, Y, Z, cmap='coolwarm')
plt.grid()
plt.axis('equal')
plt.show()
```````

```````python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
from matplotlib.ticker import LinearLocator, FormatStrFormatter

fig = plt.figure(figsize=(8,5))
ax = fig.gca(projection='3d')

X = np.arange(-5, 5, 0.25)
Y = np.arange(-5, 5, 0.25)
X, Y = np.meshgrid(X, Y)
R = np.sqrt(X**2 + Y**2)
Z = np.sin(R)

surf = ax.plot_surface(X, Y, Z, cmap='coolwarm',
                       linewidth=0, antialiased=False)

ax.set_zlim(-1.01, 1.01)
ax.zaxis.set_major_locator(LinearLocator(10))
ax.zaxis.set_major_formatter(FormatStrFormatter('%.02f'))

fig.colorbar(surf, shrink=0.5, aspect=5)

plt.show()
```````
